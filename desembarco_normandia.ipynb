La `Target_City` tiene coordenadas [**44 00 N 001 00 E**](https://share.google/pBIwTo4jbPy90rJKC) que no se corresponden con lo que estamos buscando, por tanto,  lo desechamos para nuestro estudio. 

```python
# ==========================================
# PASO 2.1: ACOTACIÓN HISTÓRICA Y ESTADÍSTICA
# ==========================================

# 1. Filtramos para quedarnos únicamente con Francia en 1944
df_normandia_1 = df_1_clean_bomb[(df_1_clean_bomb['Target Country'].str.upper() == 'FRANCE') & (df_1_clean_bomb['Mission Date'].astype(str).str.contains('1944'))]

print("--- CONTROL DE SEGMENTACIÓN (A.D.I.) ---")
print(f"Misiones totales en el dataset: {df_1_clean_bomb.shape[0]}")
print(f"Misiones identificadas en Francia durante 1944: {df_normandia_1.shape[0]}\n")

# 2. Análisis Estadístico Descriptivo de Variables Categóricas
print("--- TOP 5 CIUDADES CON MAYOR IMPACTO PATRIMONIAL ---")
print(df_normandia_1['Target City'].value_counts().head(5))

print("\n--- AVIONES MÁS UTILIZADOS EN LAS OPERACIONES ---")
print(df_normandia_1['Aircraft Series'].value_counts().head(5))

# 3. Análisis Estadístico Descriptivo de Variables Numéricas
print("\n--- ANÁLISIS DE CARGA DE IMPACTO (TONELADAS) ---")
print(df_normandia_1['Total Weight (Tons)'].describe())
```

## DATASET 2: df_2_operations

Este dataset nos muestra 15.270 misiones realizadas en Francia. 

En este dataset encontramos dos cosas llamativas:

* Impacto Geográfico

1. **El Impacto Geográfico (Variables Categóricas)**: 
    Brest (459) y St. Lô (404): Encabezan el ranking. Para tu empresa de patrimonio, esto es la justificación científica de inversión. Le dirás al comité: «Los datos demuestran que las misiones se concentraron críticamente en Brest y Saint-Lô; por ende, nuestros esfuerzos de escaneo 3D y arqueología militar de búnkeres prioritarios deben establecerse allí».
2. **Anomalía / Limpieza profunda**: Te aparece la ciudad UNKNOWN con 299 registros. Esto es genial para documentar que el dataset real tiene "ruido" y que tu software deberá filtrar los nombres desconocidos en la visualización.

* Anomalías detectadas

    * **El "Anómalo" de las Toneladas (Variables Numéricas)**: El valor máximo (`max`) de toneladas tiradas en una misión es 999.0 toneladas. Esto es un "outlier" (un valor extremo) o muy probablemente un código de error de registro en la época (un "placeholder" típico de la informática antigua cuando no sabían el dato exacto).
    * La Media es 46.7 toneladas, pero la Mediana (el 50%) es 33 toneladas. Al ser la media bastante mayor que la mediana, demuestras que hay misiones masivas muy destructivas que empujan el promedio hacia arriba.

```python
# ==========================================
# PASO 2.1: ACOTACIÓN HISTÓRICA Y ESTADÍSTICA
# ==========================================

# 1. Filtramos para quedarnos únicamente con Francia en 1944
df_normandia_2 = df_2_operations[(df_2_operations['Target Country'].str.upper() == 'FRANCE') & (df_2_operations['Mission Date'].astype(str).str.contains('1944'))]

print("--- CONTROL DE SEGMENTACIÓN (A.D.I.) ---")
print(f"Misiones totales en el dataset: {df_2_operations.shape[0]:,}")
print(f"Misiones identificadas en Francia durante 1944: {df_normandia_2.shape[0]:,}\n")

# 2. Análisis Estadístico Descriptivo de Variables Categóricas
print("--- TOP 5 CIUDADES CON MAYOR IMPACTO PATRIMONIAL ---")
print(df_normandia_2['Target City'].value_counts().head(5))

print("\n--- AVIONES MÁS UTILIZADOS EN LAS OPERACIONES ---")
print(df_normandia_2['Aircraft Series'].value_counts().head(5))

# 3. Análisis Estadístico Descriptivo de Variables Numéricas
print("\n--- ANÁLISIS DE CARGA DE IMPACTO (TONELADAS) ---")
print(df_normandia_2['Total Weight (Tons)'].describe())
```

## DATASET 3: df_3_thor_aircraft

Utilizamos este dataset para "traducir" la columna `Aircraft Series` del dataset inicial **2_operations.csv** y poder tener los nombres completos de los modelos de avión.

```python
# ==========================================================
# PASO 2.2: ENRIQUECIMIENTO DE DATOS (A.D.I. CORE)
# ==========================================================

# 1. Cargamos el glosario de aviones
# df_3_thor_aircraft = pd.read_csv("3_thor_wwii_aircraft_gloss.csv")

# 2. Fusionamos nuestro dataset de Francia con el glosario usando como puente la columna del avión
# En df_normandia_2 se llama 'Aircraft Series' y en el glosario se llama 'aircraft'
df_enriquecido = pd.merge(
    df_normandia_2, # nuestro dataset acotado a Francia 1944
    df_3_thor_aircraft, # el glosario de aviones
    left_on='Aircraft Series', # la columna del avión en nuestro dataset
    right_on='aircraft', # la columna del avión en el glosario
    how='left' # left join para no perder misiones aunque no encontremos el avión en el glosario
)

# 3. Creamos una columna limpia para el nombre del avión (si no encuentra el glosario, deja el original)
df_enriquecido['Avion_Nombre_Completo'] = df_enriquecido['full_name'].fillna(df_enriquecido['Aircraft Series'])

# 4. EXPORTACIÓN CRÍTICA: Guardamos el archivo final que alimentará Power BI y Streamlit
df_enriquecido.to_csv("adi_normandia_final.csv", index=False)

print("🚀 ¡Fusión completada con éxito por A.D.I.!")
print(f"Archivo 'adi_normandia_final.csv' generado listo para maquetar.")
```

## DATASET 5: df_5_thor_weapon

Glosario de Armamento

```python
# ==========================================================
# PASO 2.3: ENRIQUECIMIENTO FINAL CON EL GLOSARIO DE ARMAS
# ==========================================================

# 1. Cargamos tu archivo estrella de Normandía (el que guardamos en el Paso 2.2)
df_normandia = pd.read_csv("adi_normandia_final.csv")

# 2. Cargamos el diccionario de armas (Dataset 5)
df_5_thor_weapon = pd.read_csv("data/5_thor_wwii_weapon_gloss.csv")

print("--- CONEXIÓN DE ARCHIVOS REALES ---")
print(f"Misiones en Normandía cargadas: {df_normandia.shape[0]}")
print(f"Tipos de armas en el glosario: {df_5_thor_weapon.shape[0]}\n")

# 3. LA FUSIÓN: Cruzamos tus misiones con el glosario de armas
# En tu archivo se llama 'High Explosives Type' y en el glosario 'weapon_name'
df_master = pd.merge(
    df_normandia,
    df_5_thor_weapon[['weapon_name', 'weapon_type', 'weapon_class', 'weapon_description']],
    left_on='High Explosives Type',
    right_on='weapon_name',
    how='left'  # Para no borrar ninguna misión aunque la bomba no esté en el glosario
)

# 4. LIMPIEZA: Si alguna misión no encuentra su bomba en el glosario, le ponemos un texto genérico
df_master['weapon_type'] = df_master['weapon_type'].fillna('HE (High Explosive)')
df_master['weapon_description'] = df_master['weapon_description'].fillna('Bomba estándar de demolición aérea de alta intensidad')

# 5. EXPORTACIÓN MAESTRA: Guardamos el archivo FINAL ABSOLUTO de tu proyecto
df_master.to_csv("adi_dataset_master.csv", index=False)

print("🏆 ¡ENHORABUENA ELENA! TODO EL MOTOR DE DATOS ESTÁ CERRADO.")
print(f"Se ha creado 'adi_dataset_master.csv' con {df_master.shape[0]} filas.")
print("Este único archivo contiene: Francia 1944 + Nombre de Aviones + Descripción de Bombas.")
```
